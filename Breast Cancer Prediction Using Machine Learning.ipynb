```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
x = data.iloc[:,2:32]
y = data.iloc[:,1]



from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)

#KNN
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=3)
knn = knn.fit(x_train,y_train)


print("Accuracy of KNN for Train set: {:.2f}".format(knn.score(x_train,y_train)))
print("Accuracy of KNN for Test set: {:.2f}".format(knn.score(x_test,y_test)))


#Logistic Regression

from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()

lr = lr.fit(x_train,y_train)

print("Accuracy of LRN for Train set: {:.2f}".format(lr.score(x_train,y_train)))
print("Accuracy of LRN for Test set: {:.2f}".format(lr.score(x_test,y_test)))

```

```python
# decision tree

from sklearn.tree import DecisionTreeClassifier

tree = DecisionTreeClassifier()

tree = tree.fit(x_train,y_train)

print("Accuracy of Decision Tree for Train set: {:.2f}".format(tree.score(x_train,y_train)))
print("Accuracy of Decision Tree for Test set: {:.2f}".format(tree.score(x_test,y_test)))

print("Feauture importance: ",tree.feature_importances_)
```

```python
# Random Forest

from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=100,random_state=42)
rf = rf.fit(x_train,y_train)

print("Accuracy of Random Forest for Train set: {:.2f}".format(rf.score(x_train,y_train)))
print("Accuracy of Random Forest for Test set: {:.2f}".format(rf.score(x_test,y_test)))
```

```python

from sklearn.svm import SVC

svc = SVC()

svc = svc.fit(x_train,y_train)

print("Accuracy of SVM without scalar transform for Train set: {:.2f}".format(svc.score(x_train,y_train)))
print("Accuracy of SVM without scalar transform for Test set: {:.2f}".format(svc.score(x_test,y_test)))

```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.fit_transform(x_test)

svc_scalar = SVC(C=100)
scv_scalar = svc_scalar.fit(x_train_scaled,y_train)

print("Accuracy of SVM with scalar transform for Train set: {:.2f}".format(svc_scalar.score(x_train_scaled,y_train)))
print("Accuracy of SVM with scalar transform for Test set: {:.2f}".format(svc_scalar.score(x_test_scaled,y_test)))
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.fit_transform(x_test)
mlp = MLPClassifier(max_iter=1000,random_state=0)
mlp.fit(x_train_scaled, y_train)
print("Accuracy on training set for Deep learning: {:.3f}".format(mlp.score(x_train_scaled, y_train)))
print("Accuracy on test set  for Deep learning :{:.3f}".format(mlp.score(x_test_scaled, y_test)))
```
